SORU 1:
Bir hastanede bir yıl boyunca Covid olduğu bilinen 6.800 hastaya yapılan testlerin 5.700'ü pozitif çıkmıştır. Aynı süre boyunca Covid olmadığı bilinen 3.200 hastaya yapılan testlerin 2.300'ü negatif çıkmıştır.

Buna göre bu hastaneye Covid testi için giden bir kişinin testi negatif çıktıysa gerçekten Covid olmama olasılığı nedir?


Bu soru için örnek uzay hastaneden yapılan tüm testlerdir.
Soruda verilen bilgiler aşağıdaki tabloda özetlenmiştir.
image.png

A: Bir hastanın gerçekten Covid olma olayı

B: Bir hastanın testinin pozitif çıkma olayı

Soruda bizden istenen test sonucunun negatif çıktığı bilinen bir hastanın gerçekten Covid olmama olasılığıdır, bunu da koşullu olasılık olarak aşağıdaki şekilde ifade edebiliriz.

image.png

```python
from math import factorial

s_A = 1100
s_A_n = 2300
s_B_n = s_A + s_A_n
result = s_A_n / s_B_n
print("test sonucunun negatif çıktığı bilinen bir hastanın gerçekten Covid olmama olasılığı  {:.5f}".format(result))
```
